```python

# 📊 Predictive Analysis using Real Dataset
# Boston Housing Dataset


# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

# Step 2: Load dataset
df = pd.read_csv("housing.csv")

print("Dataset Loaded Successfully\n")
print(df.head())

# Step 3: Select features and target
X = df[["rm", "lstat"]]   # Input features
y = df["medv"]            # Target (house price)

# Step 4: Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Step 5: Train model
model = LinearRegression()
model.fit(X_train, y_train)

print("\nModel trained successfully!")

# Step 6: Predict
predictions = model.predict(X_test)

print("\nPredictions:\n", predictions)

# Step 7: Check error
error = mean_absolute_error(y_test, predictions)
print("\nMean Absolute Error:", error)


# 📊 GRAPH

# Take first 10 values for clarity
y_test_small = y_test.values[:10]
pred_small = predictions[:10]

x = np.arange(len(y_test_small))

plt.figure(figsize=(10,5))

# Bars
plt.bar(x - 0.2, y_test_small, width=0.4, label="Actual Price")
plt.bar(x + 0.2, pred_small, width=0.4, label="Predicted Price")

# Labels
plt.xlabel("Sample Data Points")
plt.ylabel("House Price")
plt.title("Actual vs Predicted House Prices")

plt.legend()
plt.grid(True)

plt.show()
```

## Objective
To predict house prices using machine learning.

## Dataset
Boston Housing dataset containing real housing data.

## Tools Used
Python, Scikit-learn, Google Colab

## Methodology
Used Linear Regression model to predict house prices based on features like number of rooms and poverty level.

## Result
### **The model successfully predicted house prices with minimal error.**
